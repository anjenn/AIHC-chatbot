# 05 Evaluation and Failure Analysis

Load experiment outputs, summarize top-1 and top-3 performance, and inspect failure cases.

```python
import pandas as pd

from src.config import OUTPUT_DIR
from src.evaluation import show_failures, summarize_results

baseline_results = pd.read_csv(OUTPUT_DIR / "baseline_results.csv")
knowledge_results = pd.read_csv(OUTPUT_DIR / "knowledge_results.csv")
self_consistency_results = pd.read_csv(OUTPUT_DIR / "self_consistency_results.csv")

summary_rows = []
for method_name, result_df in [
    ("baseline", baseline_results),
    ("knowledge", knowledge_results),
    ("self_consistency", self_consistency_results),
]:
    summary = summarize_results(result_df).iloc[0].to_dict()
    summary_rows.append({"method": method_name, **summary})

summary_df = pd.DataFrame(summary_rows)
summary_df.to_csv(OUTPUT_DIR / "method_summary.csv", index=False)

print(summary_df)
show_failures(self_consistency_results, n=10)
```
